```python
# split the train.csv into train and validation
import pandas as pd

# Load the dataset
df = pd.read_csv('./dataset/train.csv')

# Split the dataset into training and validation sets
train_df = df.sample(frac=0.99, random_state=42)
validation_df = df.drop(train_df.index)

# Save the training and validation sets to CSV files
train_df.to_csv('train_data.csv', index=False)
validation_df.to_csv('validation_data.csv', index=False)
```

```python
# Load the training data
train_df = pd.read_csv('train_data.csv')

# Select a subset of 100 - 200 examples
subset_df = train_df.sample(n=150, random_state=42)  # Adjust the number as needed

# Save the subset to a new CSV file
subset_df.to_csv('train_data_subset.csv', index=False)
```
